##For


```python
#For Range
for contador in range(10):
  print(contador)
```

```python
#For Range + Break
for contador in range(1000):
  print(contador)
  if contador == 10:
    break
```

##While

```python
val = 1
while (val < 6):
  val *= 2
  print(val)
```

```python
#While com else
val = 1
while val < 6:
  val *= 2

else:
  print('Valor de val é: ', val)
```
