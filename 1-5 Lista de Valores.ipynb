##Lista, tuplas e dicionários

```python
# Lista
[1,2,3,4,5]
```

```python
# Tupla
(1,2,3,4,5)
```

```python
# Dicionário
{'num1': 1, 'num2': 2}
```

##Lista

```python
lista = [34, 'Raquel', 1990]
print(lista[2])
```

##Tupla

```python
tupla = (34, 'Raquel', 1990)
print(tupla[2])
```

##Dicionário

```python
dicionario = {'nome': 'Raquel', 'cidade': 'Jundiaí'}
print(dicionario['nome'])
```

```python
dicionario['idade'] = 34
print(dicionario)
```
