## Condicional

```python
#if
if 5 < 10:
  print('Verdade, cindo é menor que dez')
```

```python
#if If/Else
num1 = 10
num2 = 5
if num1 > num2:
  print('O num1 é maior que o num2')
else:
  print('O num2 é maior que o num1')
```

```python
# If/Elif/Else
num1 = 10
num2 = 10
if num1 > num2:
  print('O num1 é maior que o num2')
elif num1 < num2:
  print('O num1 é menor que o num2')
else:
  print('Os dois numeros são iguais')
```

##Condicional com operadores lógicos

```python
num1 = 10
num2 = 5
num3 = 2
if (num1 > num2) and (num1 > num3):
  print('O maior numero é o num1')
else:
  print('O maior numero não é o num1')
```
